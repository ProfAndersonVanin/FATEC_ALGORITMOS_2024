# **Coleções de dados**
Coleção é uma estrutura que permite armazenar **múltiplos itens ou objetos**, dentro de uma única unidade. As operações que normalmente podem ser feitas em coleções são: adição, remoção, acesso, pesquisa, ordenação entre outros.

# **EXEMPLO LISTAS**

image.png

```python
# Exemplo de criação e manipulação de uma lista
frutas = ['maçã', 'banana', 'laranja', 'morango']
```

```python
frutas
```

```python
# Acessando elementos da lista
print("A primeira fruta é:", frutas[0])
print("A última fruta é:", frutas[-1])
```

```python
# Adicionando um elemento à lista
frutas.append('uva')
```

```python
frutas
```

```python
# Removendo um elemento da lista
frutas.remove('banana')
```

```python
frutas
```

```python
# Modificando um elemento da lista
frutas[2] = 'pêssego'
```

```python
frutas
```

```python
# Iterando sobre os elementos da lista
print("Todas as frutas:")
for fruta in frutas:
    print(fruta)
```

# **EXEMPLO TUPLAS**

```python
# Exemplo de criação e manipulação de uma tupla
cores = ('vermelho', 'verde', 'azul', 'amarelo')
```

```python
cores
```

```python
# Acessando elementos da tupla
print("A primeira cor é:", cores[0])
print("A última cor é:", cores[-1])
```

```python
# Iterando sobre os elementos da tupla
print("Todas as cores:")
for cor in cores:
    print(cor)
```

# **EXEMPLO DICIONÁRIOS**

image.png

```python
# Exemplo de criação e manipulação de um dicionário
pessoa = {'nome': 'João', 'idade': 30, 'cidade': 'São Paulo'}
```

```python
pessoa
```

```python
# Acessando valores do dicionário
print("Nome:", pessoa['nome'])
print("Idade:", pessoa['idade'])
print("Cidade:", pessoa['cidade'])
```

```python
# Adicionando um novo par chave-valor ao dicionário
pessoa['profissão'] = 'Engenheiro'
```

```python
pessoa
```

```python
# Modificando o valor de uma chave existente no dicionário
pessoa['idade'] = 35
```

```python
pessoa
```

```python
# Removendo um par chave-valor do dicionário
del pessoa['cidade']
```

```python
pessoa
```

```python
# Iterando sobre as chaves do dicionário
print("Informações da pessoa:")
for chave in pessoa:
    print(chave + ":", pessoa[chave])
```
